```python
import duckdb
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "browser"  # opens plots in browser tab, avoids nbformat issues

df = duckdb.sql("""
    SELECT timestamp, arrival_delay_s, precipitation
    FROM '../data/dataset_705_echandens.parquet'
    WHERE arrival_delay_s IS NOT NULL
""").df()
df["timestamp"] = pd.to_datetime(df["timestamp"])
df["hour"]      = df["timestamp"].dt.hour + df["timestamp"].dt.minute / 60
df["hour_int"]  = df["timestamp"].dt.hour

DELAY_CAP = 600
df_plot = df[df["arrival_delay_s"] <= DELAY_CAP].copy()

# ── Plot 1: 3D scatter — every trip as a point ───────────────────────────────
fig1 = go.Figure(go.Scatter3d(
    x=df_plot["hour"],
    y=df_plot["precipitation"],
    z=df_plot["arrival_delay_s"],
    mode="markers",
    marker=dict(
        size=2,
        color=df_plot["arrival_delay_s"],
        colorscale="RdYlGn_r",
        colorbar=dict(title="Delay (s)"),
        opacity=0.5,
    ),
    hovertemplate="Hour: %{x:.1f}<br>Rain: %{y:.2f} mm/h<br>Delay: %{z}s<extra></extra>",
))
fig1.update_layout(
    title="Arrival delay vs hour-of-day and precipitation (raw trips, capped at 600s)",
    scene=dict(
        xaxis_title="Hour of day",
        yaxis_title="Precipitation (mm/h)",
        zaxis_title="Arrival delay (s)",
    ),
    width=900, height=700,
)
fig1.show()

# ── Plot 2: surface — mean delay per (hour × precipitation bucket) ────────────
precip_edges  = [-0.01, 0.05, 0.5, 2.0, 5.0, 100.0]
precip_labels = ["Dry", "Drizzle", "Light", "Moderate", "Heavy"]

df_plot["precip_bin"] = pd.cut(
    df_plot["precipitation"],
    bins=precip_edges,
    labels=precip_labels,
)

pivot = (
    df_plot
    .groupby(["hour_int", "precip_bin"], observed=True)["arrival_delay_s"]
    .mean()
    .unstack("precip_bin")
    .reindex(columns=precip_labels)
)
pivot = pivot.ffill(axis=1).bfill(axis=1)

fig2 = go.Figure(go.Surface(
    x=pivot.index.tolist(),
    y=list(range(len(precip_labels))),
    z=pivot.values.T,
    colorscale="RdYlGn_r",
    colorbar=dict(title="Mean delay (s)"),
    hovertemplate="Hour: %{x}<br>Precip: %{y}<br>Mean delay: %{z:.0f}s<extra></extra>",
))
fig2.update_layout(
    title="Mean arrival delay surface: hour-of-day × precipitation level",
    scene=dict(
        xaxis=dict(title="Hour of day", tickmode="linear", dtick=2),
        yaxis=dict(
            title="Precipitation",
            tickmode="array",
            tickvals=list(range(len(precip_labels))),
            ticktext=precip_labels,
        ),
        zaxis_title="Mean delay (s)",
    ),
    width=900, height=700,
)
fig2.show()
```

```python
import sys
sys.path.insert(0, '..')

import duckdb
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from ml.preprocessors.delay_binner import DelayBinner

# ── Load data & run PCA once ──────────────────────────────────────────────────
_df = duckdb.sql("""
    SELECT * FROM '../data/dataset_705_echandens.parquet'
    WHERE arrival_delay_s IS NOT NULL
""").df()

FEATURE_COLS = [
    "time_sin", "time_cos", "dow_sin", "dow_cos", "month_sin", "month_cos",
    "is_weekend", "is_public_holiday",
    "temperature", "precipitation", "sunshine", "humidity",
    "wind_speed", "wind_gust", "pressure", "snow_depth",
    "wind_dir_sin", "wind_dir_cos",
]

_X_scaled = StandardScaler().fit_transform(_df[FEATURE_COLS].astype(float).fillna(0).values)
_pca = PCA(n_components=3)
_X_pca = _pca.fit_transform(_X_scaled)
_ev = _pca.explained_variance_ratio_
print(f"Explained variance — PC1: {_ev[0]:.1%}  PC2: {_ev[1]:.1%}  PC3: {_ev[2]:.1%}  total: {sum(_ev):.1%}")

_df_pca = pd.DataFrame(_X_pca, columns=["PC1", "PC2", "PC3"])
_df_pca["delay"] = _df["arrival_delay_s"].values

_PALETTE = ["#2196F3", "#4CAF50", "#FF9800", "#F44336", "#9C27B0", "#00BCD4", "#FF5722"]

# ── Controls ──────────────────────────────────────────────────────────────────
_slider_kw = dict(min=-300, max=600, step=5, continuous_update=False,
                  layout=widgets.Layout(width="420px"))

b1     = widgets.IntSlider(value=0,   description="B1 (s)", **_slider_kw)
b2     = widgets.IntSlider(value=60,  description="B2 (s)", **_slider_kw)
b3     = widgets.IntSlider(value=120, description="B3 (s)", **_slider_kw)
use_b2 = widgets.Checkbox(value=True,  description="enable", indent=False, layout=widgets.Layout(width="80px"))
use_b3 = widgets.Checkbox(value=True,  description="enable", indent=False, layout=widgets.Layout(width="80px"))

_out = widgets.Output()

def _render(bins):
    binner = DelayBinner(bins=bins)
    cls_series = binner.encode(_df["arrival_delay_s"]).values
    n_classes = len(bins) + 1

    fig = go.Figure()
    for cls in range(n_classes):
        mask = cls_series == cls
        sub  = _df_pca[mask]
        label = f"Class {cls}: {binner.class_names[cls]}"
        fig.add_trace(go.Scatter3d(
            x=sub["PC1"], y=sub["PC2"], z=sub["PC3"],
            mode="markers",
            name=label,
            marker=dict(size=2, color=_PALETTE[cls % len(_PALETTE)], opacity=0.45),
            customdata=sub["delay"],
            hovertemplate=(
                f"{label}<br>PC1: %{{x:.2f}}<br>PC2: %{{y:.2f}}<br>"
                f"PC3: %{{z:.2f}}<br>Delay: %{{customdata}}s<extra></extra>"
            ),
        ))
    fig.update_layout(
        title=(
            f"PCA 3D projection — colored by delay class  [bins={bins}]<br>"
            f"<sup>PC1 {_ev[0]:.1%} · PC2 {_ev[1]:.1%} · PC3 {_ev[2]:.1%} · total {sum(_ev):.1%}</sup>"
        ),
        scene=dict(
            xaxis_title=f"PC1 ({_ev[0]:.1%})",
            yaxis_title=f"PC2 ({_ev[1]:.1%})",
            zaxis_title=f"PC3 ({_ev[2]:.1%})",
        ),
        legend=dict(title="Delay class", itemsizing="constant"),
        width=1100, height=750,
    )
    _out.clear_output(wait=True)
    with _out:
        fig.show(renderer="notebook_connected")

def _on_change(_=None):
    bins = [b1.value]
    if use_b2.value: bins.append(b2.value)
    if use_b3.value: bins.append(b3.value)
    _render(sorted(set(bins)))

for _w in [b1, b2, b3, use_b2, use_b3]:
    _w.observe(_on_change, names="value")

_render([0, 60, 120])

display(widgets.VBox([
    widgets.HTML("<b>Delay class boundaries (seconds) — drag sliders or toggle to add/remove a boundary:</b>"),
    widgets.HBox([b1]),
    widgets.HBox([use_b2, b2]),
    widgets.HBox([use_b3, b3]),
    _out,
]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data/2025-01-01_istdaten.csv", sep=";")
```

```python
test_df = df[df["BETREIBER_NAME"].str.contains("MBC") & df["LINIEN_TEXT"].str.contains("705")]
test_df.head(10)
```

```python
test_df["DELAY"] = abs(pd.to_datetime(test_df["AB_PROGNOSE"]) - pd.to_datetime(test_df["ABFAHRTSZEIT"]))
test_df.head(10)
```

```python
MAX_DELAY_TRESHOLD = pd.Timedelta(minutes=1.5)
test_df["IS_DELAYED"] = False
test_df.loc[pd.to_timedelta(test_df["DELAY"]) >= MAX_DELAY_TRESHOLD, "IS_DELAYED"] = True
print(f"705 is late {len(test_df["IS_DELAYED"] == True) / len(test_df)}% of time")
print(test_df["IS_DELAYED"] == True)
```
